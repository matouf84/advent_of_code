```python

fn = "01-input.txt"
with open(fn) as df:
    current_sum = 0
    max_calories = 0
    lines = df.readlines()
    for c in lines:
        # print(c, end=": ")
        if len(c.strip())==0:
            max_calories = max(max_calories, current_sum)
            # print("max_calories = ", max_calories)
            current_sum = 0
        else:
            current_sum += int(c)
            # print("current_sum = ", current_sum)

    print("Max calories:", max_calories)
```

```python

import numpy as np

fn = "01-input.txt"
with open(fn) as df:
    current_sum = 0
    max_calories = 0
    max_calories_list = []
    lines = df.readlines()
    for i, c in enumerate(lines):
        # print(i, ": value = ", c.strip(), end=" -> ")
        current = int(c.strip() or 0)
        current_sum += current
        # print("current_sum = ", current_sum, end=" -> ")
        if (current == 0) or (i == len(lines)-1):
            # max_calories = max(max_calories, current_sum)
            # print("max_calories = ", max_calories)
            max_calories_list.append(current_sum)
            print("max_calories_list = ", max_calories_list )
            current_sum = 0
        # print()
        
    max_calories_list.sort(reverse=True)
    print(max_calories_list)
    top_3 = np.sum(max_calories_list[:3])
    print("Top 3 Sum:", top_3)
```
